### 2. Нахождения коэффициентов регрессии с попощью МНК

###### Реализуем функции необходимые для нахождения коэффициентов регрессии:

```python
#Формирование матрицы H по столбцам датафрейма data
def get_H(data, features_lst):
    feature_arr = np.empty((data.shape[0], len(features_lst) + 1))
    feature_arr[:, 0] = 1
    feature_arr[:, 1:] = data[features_lst]
    return feature_arr

#Формирование матрицы H для заданной степени полинома - poly_degree
def get_poly(data, poly_degree):
    poly = np.zeros((data.size, poly_degree + 1))
    for col_num in range(poly_degree + 1):
        poly[:, col_num] = data**col_num
    return poly

#Вычисление суммы квадратов ошибок
def getRSS(H, y, w):
    return (y - H.dot(w)).T.dot(y - H.dot(w))

#Вычисление среднеквадратичной ошибки
def getMSE(H, y, w):
    return getRSS(H, y, w) / y.shape[0]

#Нахождение вектора весов w с помощюь формулы МНК в матричном виде 
def get_w_simple(H, y):
    return linalg.inv(H.T.dot(H)).dot(H.T).dot(y)

#Вычисление среднеквадратичной ошибки для 5 фолдов
def get_avg_MSE_simple(H, y):
    kf = KFold(n_splits=5)
    MSE_sum = 0
    for train, test in kf.split(y):
        pred_koeff = get_w_simple(H[train, :], y[train])
        MSE_sum += getRSS(H[test, :], y[test], pred_koeff) / y[test].shape[0]
    return MSE_sum / 5
```

###### Средняя среднеквардатичная ошибка для регрессии на 'Humidity', 'Wind_Speed', 'Visibility':

```python
H = get_H(data, ['Humidity', 'Wind_Speed', 'Visibility'])
y = np.array(data['Temp'])
MSE1 = get_avg_MSE_simple(H, y)
MSE1
```

### 3. Подбор степеней полиномиальной регрессии вручную

```python
#Функция подбирает лучшую степень полинома и выводит график зависимости степени полинома от точности
def best_degree(x, y, max_poly_degree, ax, name):
    kf = KFold(n_splits=5)
    avg_MSE = []
    for poly_degree in range(1, max_poly_degree + 1, 1):
        MSE_sum = 0
        for train, test in kf.split(x):
            H_train = get_poly(x[train], poly_degree)
            pred_koeff = get_w_simple(H_train, y[train])
            H_test = get_poly(x[test], poly_degree)
            MSE_sum += getRSS(H_test, y[test], pred_koeff) / y[test].shape[0] 
        avg_MSE.append(MSE_sum / 5)
    min_avg_MSE_idx = np.argmin(avg_MSE)
    ax.set_xlabel('poly degree')
    ax.set_ylabel('MSE')
    ax.set_title(
        name +
        ' (best degree - ' + str(min_avg_MSE_idx + 1) + 
        ', MSE - ' + str(round(avg_MSE[min_avg_MSE_idx], 2)) + ')')
    ax.plot(range(1, len(avg_MSE) + 1, 1), avg_MSE);
```

###### Графики зависимости ошибки от степени полинома:

```python
y = np.array(data['Temp'])
figure, axes = plt.subplots(2, 2, figsize=(20,10))
best_degree(data['Humidity'], y, 15, axes[(0,0)], 'Humidity')
best_degree(data['Wind_Speed'], y, 6, axes[(0,1)], 'Wind_Speed')
best_degree(data['Visibility'], y, 11, axes[(1,0)], 'Visibility')
```

###### По полученным данным стараемся подобрать оптимальные степени полиномов:

### 4. Ридж-регрессия:

```python
#Нахождения вектора весов для заданной ламбды - lam
def get_w_ridge(H, y, lam):
    I_mod = np.ones((H.shape[1], H.shape[1]))
    I_mod[0,0] = 0 
    return linalg.inv(H.T.dot(H) + lam * I_mod).dot(H.T).dot(y)

def get_avg_MSE_ridge(H, y, lam):
    kf = KFold(n_splits=5)
    MSE_sum = 0
    for train, test in kf.split(y):
        pred_koeff = get_w_ridge(H[train, :], y[train], lam)
        MSE_sum += getRSS(H[test, :], y[test], pred_koeff) / y[test].shape[0]
    return MSE_sum / 5
```

```python
poly_degree_lst = list(range(1, 11, 1))
best_lam_lst = []
best_MSE_lst = []
for poly_degree in poly_degree_lst:
    H = np.concatenate((
        get_poly(data['Humidity'], poly_degree), 
        get_poly(data['Wind_Speed'], poly_degree)[:, 1:],
        get_poly(data['Visibility'], poly_degree)[:, 1:]), axis=1)
    y = np.array(data['Temp'])
    lam_arr = np.logspace(-3, 4, num=8)
    avg_MSE_lst = []
    for lam in lam_arr:
        avg_MSE_lst.append(get_avg_MSE_ridge(H, y, lam))
    min_idx = np.argmin(avg_MSE_lst)
    best_lam_lst.append(lam_arr[min_idx])
    best_MSE_lst.append(avg_MSE_lst[min_idx])

min_idx = np.argmin(best_MSE_lst)
figure, axes = plt.subplots(1, 1, sharex=True, sharey=False, figsize=(20,10))
axes.annotate(
    'lam = ' + str(best_lam_lst[min_idx]) + 
    ', MSE = ' + str(round(best_MSE_lst[min_idx], 2)), 
    (poly_degree_lst[min_idx], best_MSE_lst[min_idx]), 
    fontsize=20)
axes.set_xlabel('poly degree')
axes.set_ylabel('MSE')
axes.scatter(poly_degree_lst, best_MSE_lst, marker='o');
```

### 5. Градиентный спуск

```python
# Расчет батч-градиента
def get_gradient_batch(H, y, w):
    return -2 * H.T.dot(y - H.dot(w))

# Расчет стохастического градиента
def get_gradient_stoch(H, y, w):
    idx = randint(0, y.shape[0] - 1)
    return -2 * H[idx, :] * (y[idx] - H[idx, :].dot(w))

# Расчет мини-батч градиента
def get_gradient_mini_batch(H, y, w):
    batch_size = 1000
    rand_idx = randint(0, y.shape[0] - 1)
    if rand_idx + batch_size > y.shape[0]:
        H_batch = H[rand_idx - batch_size:rand_idx, :]
        y_batch = y[rand_idx - batch_size:rand_idx]
    else:
        H_batch = H[rand_idx:rand_idx + batch_size, :]
        y_batch = y[rand_idx:rand_idx + batch_size]
    return -2 * H_batch.T.dot(y_batch - H_batch.dot(w))

# Вычисление магнитуды(длины) градиента
def get_magnitude(gradient):
    return math.sqrt(sum(gradient**2))

# Реализация градиентного спуска
def regression_gradient_descent(H, y, w0, step_size, tolerance, get_gradient, max_steps):
    w = w0
    gradient = get_gradient(H, y, w)
    n_steps = 0
    while get_magnitude(gradient) > tolerance:
        if n_steps > max_steps:
            break
        n_steps += 1
        w = w - step_size * gradient
        gradient = get_gradient(H, y, w)
    return (w, n_steps)
```

```python
def show_plot_tolerance_mse(tolerance_lst, MSE_lst, steps_lst, x_scale):
    figure, axes = plt.subplots(1, 1, sharex=True, sharey=False, figsize=(20,10))
    for i, annot in enumerate(steps_lst):
        axes.annotate(
            'steps - ' + str(round(annot, 2)), 
            (tolerance_lst[i], MSE_lst[i]))
    axes.set_xscale(x_scale)
    axes.set_xlabel('tolerance')
    axes.set_ylabel('MSE')
    axes.scatter(tolerance_lst, MSE_lst, marker='o');
```
