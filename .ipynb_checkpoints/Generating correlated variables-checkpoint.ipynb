```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

###### Первым делом приходит на ум к сгенерированной величине добавить шум, и его силу регулировать с помощью множетеля:

```python
_, axes = plt.subplots(4, 5, sharex=True, sharey=True, figsize=(20,10))
n = 1000
k = 0.0
for idx in range(20):
    k += 0.1
    x = np.random.normal(size=n)
    y = x + k * np.random.normal(size=n)
    axes[int(idx / 5), idx % 5].set_title(round(k,1))
    axes[int(idx / 5), idx % 5].scatter(x, y)
```

###### Можно воспользоваться формулой из интернета, где можно точно регулировать коэффициент корреляции:

```python
_, axes = plt.subplots(2, 5, sharex=True, sharey=True, figsize=(20,10))
n = 1000
k = 0.0
for idx in range(10):
    k += 0.1
    x = np.random.normal(size=n)
    y = k * x + math.sqrt(1.0 - k**2) * np.random.normal(size=n)
    axes[int(idx / 5), idx % 5].set_title(round(k,1))
    axes[int(idx / 5), idx % 5].scatter(x, y)
```

###### Для нелинейного случая можно также добавлять шум:

```python
_, axes = plt.subplots(4, 5, sharex=True, sharey=True, figsize=(20,10))
n = 1000
k = 0.0
for idx in range(20):
    k += 0.1
    x = np.random.normal(size=n)
    y = x**2 + k*np.random.normal(size=n)
    axes[int(idx / 5), idx % 5].set_title(round(k,1))
    axes[int(idx / 5), idx % 5].scatter(x, y)
```
